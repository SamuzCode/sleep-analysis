## Analisis de las ocupaciones

```python
data.groupby('Occupation')['Gender'].value_counts()
```

```python
print(data['Sleep Duration'].min(), data['Sleep Duration'].max())
```

```python
data.groupby('Gender')[['Sleep Duration','Quality of Sleep']].mean()
```

```python
#Podenoms en una variable para poder hacer un top o oragnizarlos si no ahi ascending false es de manor a mayor y si no es true 
#o si simplementio no esta ya que esta predeterminadamente esta en true

#en data agrupamos por ocupacion en forma de las menjor duracion y el calidad de sueño y sacar el medio (mean)
#calidad y duracion del sueños en las distintas ocupaciones

occ_qua= data.groupby('Occupation')[['Sleep Duration','Quality of Sleep']].mean()
occ_qua.sort_values(by='Quality of Sleep', ascending=False)
```

```python
#El indice de maasa corporal por ocupaciones
#como BMI CATEGORY es string no podemos usar mean o min o max entonces ponemos value_counts

occ_bmi = data.groupby(['Occupation','Gender'])['BMI Category'].value_counts()
occ_bmi.sort_values(ascending=False)
```

```python
#ponemos como el valor en base calidad de sueño
occ_steps = data.groupby('Occupation')[['Daily Steps','Physical Activity Level','Sleep Duration','Quality of Sleep']].mean()
occ_steps.sort_values(by='Quality of Sleep',ascending=False)
```
